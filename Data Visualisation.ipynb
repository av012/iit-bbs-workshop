```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line Plot
x = np.array([1,2,3,4,5])
y = x**2

print(x)
print(y)
```

```python
plt.style.use("seaborn")
```

```python
plt.figure(figsize=(5,5))
plt.plot(x,y,color='red',marker='o',label='Prices 2019')
plt.scatter(x,x,label='Prices 2018')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price Chart")
plt.legend()
plt.show()
```
