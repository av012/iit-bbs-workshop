```python
def error(X,Y,theta):
    Y_ = getYpred(theta,X)
    m = X.shape[0]
    total_error = np.sum((Y-Y_)**2)
    return total_error/m

def getYpred(theta,X):
    return theta[1]*X + theta[0]
    
def getGrads(X,Y,theta):
    grads = np.zeros((2,))
    Y_ = getYpred(theta,X)
    m = X.shape[0]
    grads[0] = 2*np.sum(Y_-Y)
    grads[1] = 2*np.sum((Y_-Y)*X)
    
    return grads/m
```

```python
def pltGraph(X,Y,theta):
    Ypred = getYpred(theta,X)
    plt.plot(X,Ypred,color='red')
    plt.scatter(X,Y)
    plt.show()
```

```python
def gradDescent(X,Y,lr):
    theta = np.zeros((2,))
    error_list = []
    for i in range(25):
        grads = getGrads(X,Y,theta)
        theta[0] = theta[0] - lr*grads[0]
        theta[1] = theta[1] - lr*grads[1]
        e = error(X,Y,theta)
        error_list.append(e)
        if i%5==0:
            pltGraph(X,Y,theta)
        
    return theta,error_list
    
```

```python
theta,error_list = gradDescent(X,Y,0.1)
```

```python
plt.plot(error_list)
plt.show()
```
