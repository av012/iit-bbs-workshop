```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = (x-5)**2

plt.plot(x,y)
plt.show()
```

```python
import time
```

```python
#min x that minimizes y (cost or error fn)
x_ = 0
learning_rate = 0.1
error_list = []


plt.plot(x,y)
for i in range(30):
    
    x_ = x_ - learning_rate*2*(x_-5)
    error = (x_-5)**2.0 
    error_list.append(error)
    #print("X : %.2f, Error : %.4f"%(x_,error))
    plt.scatter(x_,error)
plt.show()
   
    
```

```python
plt.plot(error_list)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.show()
```
